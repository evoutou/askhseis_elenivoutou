# Gaussian Κυματοσυνάρτηση

Δεδομένης της κατάστασης ενός συστήματος για χρόνο $ t = 0 $, πώς θα εξελιχθεί το σύστημα με την πάροδο του χρόνου;  Αυτό είναι ένα από τα πιο θεμελιώδη και σημαντικά ερωτήματα που μπορεί κανείς να θέσει στη φυσική. Στην τελική, ο λόγος ύπαρξης μιας θεωρίας είναι να κάνει προβλέψεις που μπορεί να δοκιμάσει ένα πείραμα.  Στην κβαντική μηχανική η κατάσταση ενός συστήματος την χρονική στιγμή $ t = 0 $ δίνεται από την κυματομορφή, $ \Psi (x, 0) $, και για να κάνουμε προβλέψεις, χρειαζόμαστε κάποιο τρόπο για να υπολογίσουμε την κυματομορφή σε άλλες χρονικές στιγμές, $ \Psi(x, t) $.   Αυτός θα μας επιτρέψει να κάνουμε προβλέψεις όπως:
1. Πόσο πιθανό είναι το σωματίδιο να παραμείνει σε ένα συγκεκριμένο μέρος
2. Πώς ανταποκρίνεται το σύστημα σε διαταραχές
3. Πώς τα παρατηρίσιμα μεγέθη όπως η ενέργεια, η περιστροφή, η ορμή, η μαγνήτιση κλπ. εξελίσσονται χρονικά

Ένας βολικός τρόπος υπολογισμού της $ \Psi (x, t) $ από την $ \Psi (x, 0) $ για ένα ελεύθερο σωματίδιο είναι μέσω μετασχηματισμού Fourier (Griffiths 2.4, και πρόβλημα του Griffiths 2.22). Η λογική είναι η εξής - γνωρίζουμε ότι οι ενεργειακές ιδιοκαταστάσεις του ελεύθερου σωματιδίου δίνονται από επίπεδα κύματα και χαρακτηρίζονται από μια ορμή, $ k $.
$$\Psi_k(x,t) \approx \exp\left( i(kx-\frac{\hbar k^2}{2m}t)\right), \; E_k=\frac{\hbar^2 k^2}{2m}$$
Έτσι για $ t = 0 $, έχουμε ένα πλήρες σύνολο καταστάσεων, $ \Psi_k (x, 0) $, για το οποίο μπορούμε εύκολα να προσθέσουμε την εξάρτηση του χρόνου τους και υπερθέτοντας να δημιουργήσουμε οποιαδήποτε άλλη αυθαίρετη κατάσταση, $ \Psi (x, 0) $.  Εάν το βάρος της ιδιοκατάστασης $ \Psi_k = e^{ikx} $ στην υπέρθεση είναι $ \phi (k) $, τότε μπορούμε να γράψουμε την υπέρθεση ως
$$\Psi(x,0)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}\phi(k)e^{ikx}dk$$
Παίρνοντας ως δεδομένο το $\Psi(x, 0)$, ο στόχος είναι να βρούμε το $ \phi(k)$, το οποίο μπορεί να γίνει μέσω μετασχηματισμού Fourier:
$$\phi(k)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}\Psi(x,0)e^{-ikx}dx$$
Μόλις υπολογίσουμε το $\phi (k) $, το μετασχηματισμό Fourier του $ \Psi (x, 0) $, μπορούμε εύκολα να προσθέσουμε την εξάρτηση από το χρόνο κάθε ιδιοκατάστασης και να το μετασχηματίσουμε πίσω για να αποκτήσουμε το $ \Psi (x, t) $.
$$\Psi(x,t)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}\phi(k)e^{ i(kx-\frac{\hbar k^2}{2m}t)}dk$$
Έτσι, η συνταγή υπολογισμού του $\Psi(x, t)$ από το $ \Psi(x, 0) $ είναι:
1. Fourier μετασχηματισμός του $\Psi (x, 0) $ για να πάρουμε το $ \phi(k) $
2. Προσθέση της εξάρτησης χρόνου, $\phi(k,t)=\phi(k)e^{ -i\frac{\hbar k^2}{2m}t}$
3. Αντίστροφη μετασχηματισμού Fourier $ \phi (k, t) $ για να πάρουμε την $ \Psi (x, t) $

Ένα καλό παράδειγμα αυτής της διαδικασίας είναι η χρονική εξέλιξη μιας Gaussian κυματομορφής. Το Πρόβλημα 2.22 του Griffiths σας δείχνει πώς να το κάνετε αυτό αναλυτικά, και αυτό το notebook θα σας διδάξει πώς να το κάνετε υπολογιστικά.  Η Gaussian κυματομορφή παρέχει ένα καλό έδαφος δοκιμών, αλλά το ωραίο για την υπολογιστική λύση είναι ότι μπορεί να εφαρμοστεί σε οποιοδήποτε αρχική $ \Psi (x, 0) $. Αυτό είναι αρκετά σημαντικό, αφού τα ολοκληρώματα που εμπλέκονται στην αναλυτική λύση είναι συνήθως αρκετά δύσκολα, συνήθως απροσδιόριστα, για οτιδήποτε πιο περίπλοκο από μια Gaussian.


## Εξέλιξη χρόνου μιας Gaussian κυματομορφής

Η $ \Psi (x, 0) $ δίνεται από μια Gaussian,
$$ \Psi (x, 0) = ae^{-ax^2} $$
Θα ορίσουμε $ a = m = \hbar = 1 $.  Το πεδίο ορισμού για την υπολογιστική λύση θα είναι η περιοχή από $ x = -20 $ έως $ x = 20 $, με 400 διακριτά σημεία. Ξεκινάμε φορτώνοντας τις απαραίτητες βιβλιοθήκες, κάνοντας το δίκτυο σε $ x $ και $ k $, καθορίζοντας την $ \Psi (x, 0) $.

```python
%matplotlib ipympl   
import matplotlib.pyplot as plt 
import numpy    
```

```python
N=400      # Αριθμός σημείων του πεδίου ορισμού
L=20
dx = 2.*L/N
x = numpy.arange(-L,L,dx)   # Ορίστε ένα πλέγμα από x = -l έως x = l, με απόσταση 2L/n
x_exact = numpy.arange(-L,L,4*dx)  # Πιο αραιό πλέγμα, που χρησιμοποιείται μόνο στο animation, για να σχεδιάσoυμε ακριβή λύση
a=1
m=1
hbar=1
psi_0 = numpy.exp(-a*x**2)
psi_0 = psi_0 / numpy.sqrt(dx*sum(numpy.conjugate(psi_0)*psi_0))
```

### Ερώτηση 1
Στο λύση του 2.22 προβλήματος του Griffiths , βρίσκεται η ακριβής λύση για την χρονική εξέλιξη αυτής της Gaussian.  Χρησιμοποιήστε αυτή τη λύση για να ολοκληρώσετε αυτή τη συνάρτηση, η οποία παίρνει το $A$, το πλέγμα $x$, και το χρόνο $ t $ και επιστρέφει την $ \psi_ {Exact} (x, t) $.

```python
def exact_solution(a, x, t):
    a_t = 1.0 / numpy.sqrt(1.0 + 2j*a*hbar*t/m)
    
    return   a_t * numpy.exp(-a * x**2 / (1.0 + 2j*a*hbar*t/m))# Συμπληρώστε την ακριβή λύση
```

### Ερώτηση 2
Προκειμένου να εφαρμόσετε την εξάρτηση από το χρόνο της κυματομορφής, πρέπει να γνωρίζετε την ορμή, $ k $, για κάθε ιδιοκατάσταση της υπέρθεσης.  Μόνο τότε θα μπορέσετε να υπολογίσετε τον συντελεστή φάσης στην $\phi(k,t)=\phi(k)e^{ -i\frac{\hbar k^2}{2m}t}$.  Χρησιμοποιήστε τη λειτουργία fftfreq για να πάρετε μια σειρά από αυτές τις ορμές.

```python
k = 2*numpy.pi*numpy.fft.fftfreq(N, dx)  # Βρειτε τις ορμές από τη συνάρτηση FFT
```

### Ερώτηση 3
Τώρα το σημαντικότερο κομμάτι του προβλήματος. Ολοκληρώστε αυτή τη συνάρτηση που επιστρέφει  την $ \Psi (x, t) $, δεδομένων των $ k $, $ \Psi (x, 0) $, και του χρόνου $ t $. Κάνετε αυτό χρησιμοποιώντας τις συναρτήσεις fft και ifft για να υπολογίσετε το $ \phi (k) $, έπειτα το $\phi(k, t) = \phi(k) e^{-i \frac{\hbar k^2}{2m}t} $, και τελικά να μετασχηματιστεί πίσω για να πάρει την $ \Psi ( x, t) $.

Σημείωση:  Πρέπει μόνο να συμπληρώσετε τις δύο γραμμές για $ \phi (k, t) $ και $ \Psi (x, t) $. Η άλλη έχει συμπληρωθεί ήδη.

```python
def time_evolve(k, psi_x, t):
    phi_k = numpy.fft.fft(psi_x)               # Μετατρέψτε σε χώρο ορμής
    phi_k_t =   phi_k * numpy.exp(-1j * (hbar * k**2 / (2*m)) * t)    # Εφαρμόστε την εξάρτηση χρόνου
    psi_x_t =    numpy.fft.ifft(phi_k_t)  # Μετατρέψτε πίσω στο πραγματικό χώρο
    return psi_x_t
```

### Ερώτηση 4
Εντάξει, τώρα πρόκειται να χρησιμοποιήσoυμε τις συναρτήσεις για να δουμε την εξέλιξη του χρονικού κύματος της Gaussian.  Θα δούμε τα πραγματικά και φανταστικά μέρη της κυματομορφής, καθώς και την πυκνότητα πιθανότητας, $ |\psi (x, t) |^2 $.  Οι τελείες είναι η ακριβή λύση και οι γραμμές είναι η αριθμητική λύση.

Κάτω από τα γραφήματα, σχολιάστε τα εξής:
1. Οι αναλυτικές και αριθμητικές λύσεις συμφωνούν;
2. Πιστεύετε ότι το animation είναι ακριβές για όλες τις χρονικές περιοχές; Γιατί ή γιατί όχι;

```python
import matplotlib.animation as animation
  
# ΕΙσάγουμε τα 2 υπογραφήματα
fig, (ax1, ax2, ax3) = plt.subplots(3,1)

# Δημιουργούμε δύο αντικείμενα γραμμής (ένα σε κάθε άξονα)
line1, = ax1.plot([], [], lw=2)
line4, = ax1.plot([], [], 'o')
line2, = ax2.plot([], [], lw=2, color='g')
line5, = ax2.plot([], [], 'o', color='g')
line3, = ax3.plot([], [], lw=2, color='r')
line6, = ax3.plot([], [], 'o', color='r')
line = [line1, line2, line3, line4, line5, line6]

# τους ίδιους άξονες όπως πριν (μόλις το κάνουμε και για τους δύο)
ax1.set_ylim(-1,1)
ax1.set_xlim(-L, L)
ax1.grid()
ax1.set_ylabel("Re(Wavefunction)")
ax1.set_title("Evolution of Wavefunction and Probability Density in real space")

ax2.set_ylim(-1,1)
ax2.set_xlim(-L, L)
ax2.grid()
ax2.set_ylabel("Im(Wavefunction)")

ax3.set_ylim(0, 1)
ax3.set_xlim(-L, L)
ax3.grid()
ax3.set_xlabel("x")
ax3.set_ylabel("Probability Density")

time_text = ax1.text(0.05, 0.95,'',horizontalalignment='left',verticalalignment='top', transform=ax1.transAxes)
norm_text = ax1.text(0.05, 0.85,'',horizontalalignment='left',verticalalignment='top', transform=ax1.transAxes)
time_per_frame = 0.1  # Θα μπορούσατε να το αλλάξετε για να κάνετε το animation πιο γρήγορo ή πιο αργό

def run_init():
    line[0].set_data([], [])
    line[1].set_data([], [])
    line[2].set_data([], [])
    return line,
    
def run(i):
    wf = time_evolve(k, psi_0, time_per_frame*i)
    wf_exact = exact_solution(a, x_exact, time_per_frame*i)
    y1data = numpy.real(wf)
    y2data = numpy.imag(wf)
    y3data = numpy.conjugate(wf)*wf
    
    y1data_exact = numpy.real(wf_exact)
    y2data_exact = numpy.imag(wf_exact)
    y3data_exact = numpy.conjugate(wf_exact)*wf_exact

    # Ενημερώστε τα δεδομένα των αντικειμένων των τριών γραμμών
    line[0].set_data(x, y1data)
    line[1].set_data(x, y2data)
    line[2].set_data(x, y3data)
    line[3].set_data(x_exact, y1data_exact)
    line[4].set_data(x_exact, y2data_exact)
    line[5].set_data(x_exact, y3data_exact)
    # Εμφάνιση του τρέχοντος χρόνου του animation
    time_text.set_text('tau = %.2f' % (i*time_per_frame))
    norm_text.set_text('norm = %.2f' % (dx*sum(numpy.conjugate(wf)*wf)))
    return line, time_text, norm_text

ani = animation.FuncAnimation(fig, run, init_func=run_init,
                               frames=1000, interval=50, blit=True)
    
plt.show()
```
